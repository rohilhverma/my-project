```python
def game_date(x):
    z = x["game_id"].split("_")[0].split("-")
    return datetime(int(z[0]), int(z[1]), int(z[2]))
def matchup(x):
    z = np.sort([x["Team"],x["Opponent"]])
    return f"{z[0]} vs. {z[1]}"
def home_team(x):
    if (x["local"] == 0.0):
        return x["Opponent"]
    return x["Team"]
def away_team(x):
    if (x["local"] == 1.0):
        return x["Opponent"]
    return x["Team"]
def home_team_id(x):
    return team_mapping[x["home_TEAM"]][0]
def away_team_id(x):
    return team_mapping[x["away_TEAM"]][0]
def home_moneyline(x):
    if (x["local"] == 0.0):
        if (x["odds.opponent"] >= 2):
            return int(round((x["odds.opponent"] - 1) * 100))
        return int(round(-100 / (x["odds.opponent"] - 1)))
    if (x["odds"] >= 2):
        return int(round((x["odds"] - 1)*100))
    return int(round(-100 / (x["odds"] - 1)))
def away_moneyline(x):
    if (x["local"] == 1.0):
        if (x["odds.opponent"] >= 2):
            return int(round((x["odds.opponent"] - 1) * 100))
        return int(round(-100 / (x["odds.opponent"] - 1)))
    if (x["odds"] >= 2):
        return int(round((x["odds"] - 1)*100))
    return int(round(-100 / (x["odds"] - 1)))
```

```python
z["GAME_DATE"] = z.apply(game_date, axis=1)
# z["MATCHUP"] = z.apply(matchup, axis=1)
z["home_TEAM"] = z.apply(home_team, axis=1)
z["away_TEAM"] = z.apply(away_team, axis=1)
z["home_TEAM_ID"] = z.apply(home_team_id, axis=1)
z["away_TEAM_ID"] = z.apply(away_team_id, axis=1)
```
