```python
from nba_api.stats.static import teams, players
import pandas as pd
import numpy as np
import sqlite3
import os
import time

pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
conn = sqlite3.connect('nba.db')
cursor = conn.cursor()
```

```python
nba_players = players.get_active_players()
nba_players_id = {}
for x in nba_players:
    name = x['full_name']
    nba_players_id[f'{name} Id'] = x["id"]
ids = nba_players_id.values()
```

```python
cursor.execute("SELECT DISTINCT GAME_ID FROM [2020-2021 Team Data]")
game_ids = [x[0] for x in cursor.fetchall()]
```

```python
from nba_api.stats.endpoints import BoxScoreTraditionalV3
box_scores = []
for x in game_ids:
    box_scores.append(BoxScoreTraditionalV3(game_id=x))
    time.sleep(1.3)
    
```
